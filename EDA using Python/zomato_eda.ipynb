### Analysis of categorical variable:

```python
#checking how many times a value occurs in a particular column

final_df.Country.value_counts()   #wrt a particular country how many records are there

#zomato operates majorly in  india
```

```python
country_names = final_df.Country.value_counts().index   #indexing the countries based on their occurence in the dataset
```

## Grouping interrelated variables:

```python
##checking the ratings

final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text'])
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
##converting the above dataset into a dataframe

final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
#renaming the zero column and saving it in a variable

ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

## Obseravations and Conclusions:

1. When Aggregate rating is >=4.5 and <=4.9 
    * Rating text- Excellent with color- Dark Green.

2. When Aggregate rating is >=4.0 and <=4.4 
    *  Rating text- Very Good with color- Green.

3. When Aggregate rating is >=3.5 and <=3.9 
    *  Rating text- Good with color- Yellow.

4. When Aggregate rating is >=2.5 and <=3.4 
    *  Rating text- Average with color- Orange.
    
5. When Aggregate rating is >=1.8 and <=2.4 
    *  Rating text- Poor with color- Red.

## Find the country name that has given 0 rating

```python
final_df
```

```python
country_ratings = final_df.groupby(['Country', 'Aggregate rating']).size().reset_index().rename(columns={0:'Count'})
```

```python
country_ratings
```

```python
zero_rating_country = []
for rating, country in zip(country_ratings['Aggregate rating'], country_ratings['Country']):
    if rating ==0:
        zero_rating_country.append(country)
```

```python
zero_rating_country
```

```python
## another way
[country for rating, country in zip(country_ratings['Aggregate rating'], country_ratings['Country']) if rating ==0] 
```
